```python
import pandas as pd
```

```python
students = {
    'StudentID': ['1021', '1022', '1023', '1024'],
    'FirstName': ['Kemi', 'Lara', 'Mike', 'Niyi'],
    'LastName': ['Ade', 'Olu', 'Oladeji', 'Olu'],
    'Gender': ['Female', 'Female', 'Male', 'Male'],
    'Age': [17, 18, 16, 17],
    'Math': [80, 85, 90, 75],
    'Computer Science': [85, 80, 90, 75],
    'Sciences': [95, 90, 85, 92],
    'English': [92, 85, 88, 90],
    'GPA': [3.42, 3.14, 3.51, 2.95],
    'Class': ['SS2', 'SS3', 'SS1', 'SS2']
}
```

```python
Student_df = pd.DataFrame(students)
```

```python
Student_df
```

```python
Student_df.sort_values(by='LastName', ascending=False)
# Notice that we have two (olu)s in our data.
# Now we have to sort them by their FisrtName if they have the same LastName. That's how our schools sort our names right??
# To do this, we just need to pass in a list of the variables we want to use to sort them in order of their precedence.
# Check the code below to see how this is done.
```

```python
Student_df.sort_values(by=['LastName', 'FirstName'], ascending=False, inplace=True)
```

```python
# we can also specify the order we want each element of the by=[] list to be sorted
```

```python
Student_df.sort_values(by=['LastName', 'FirstName'], ascending=[False, True], inplace=True)
```

```python
# We can also just sort them out with their index
Student_df.sort_index(inplace=True)
```
