Funny enough, we can change the name of each column. You just have to re-assign it.
In the code below,  I changed the names of te first three columns. 
`{'StudentID', 'FirstName', 'LastName'}` to `{'Matrix_Number', 'First_Name', 'Last_Name'}`

```python
students_df.columns # Now, check the result of this
```

We can also perform this action with the `.rename(columns={oldname : newname, ...})` method. When we want to change some of the columns name and not all, it is preferrably to use `.rename(columns={oldname : newname, ...})`.

```python
students_df.rename(columns={'Matric_Number':'StudentID', 'First_Name':'FirstName', 'Last_Name':'LastName'}, inplace=True)
```

### Updating our data inside a data frame
#### Using `loc`
We can use `loc` to update the data in our dataframe.

```python
students_df
```

Let's change the gender of the person in row `7` from `Male` to `Female`

```python
students_df.loc[7, 'Gender'] = 'Female'
```

```python
students_df # Notice that `Martinez` is now a female 😂 😂 😂... It's not my fault oooooo......
```

# Updating multiple rows at a time...
Let's say there was a report a mass examination malpractice during all exams conducted in the school and the pricipal instructed that the exam should be re-conducted. As the IT team, we have to clear the current score in our record and put ` - ` for every student.

```python
students_df.loc[:, ['Math', 'Computer Science', 'Sciences', 'English', 'GPA']] = '-'
```
