# Advanced Updating
#### We use the following methods to perform update operations in pandas
- apply()
- map()
- applymap()
- replace()

#### 1. Apply:
The apply() method in pandas is used to apply a function along an axis of the DataFrame. This function can be a predefined function or a user-defined function. It can be applied to a single column or multiple columns of a DataFrame.

- When applied to a Series object, .apply() works similarly to .map(). The function is applied to each element of the Series     object and the output is returned as a new Series object.

- When applied to a DataFrame, the .apply() method is used to apply a function to either the rows or columns of the DataFrame, depending on the value of the axis parameter. If axis=0, the function is applied to each column of the DataFrame and the output is returned as a new DataFrame with the same columns. If axis=1, the function is applied to each row of the DataFrame and the output is returned as a new DataFrame with the same rows.

```python
students_df['FirstName'] = students_df['FirstName'].apply(str.upper) # To change all the students first_name to uppercase.
```

```python
students_df
```

```python
students_df['FirstName'] = students_df['FirstName'].apply(lambda x: x.title()) # # To change all the students first_name back to title_casing.
```

```python
students_df
```

```python
students_df.apply(len) # The dafault axis is rows, check the code below and notice the difference.
```

```python
students_df.apply(len, axis='columns')
```

```python
students_df.loc[:, "Math" : "English"].apply(pd.Series.max) # The highest mark in each subject.
```

#### 2. Applymap
The applymap() method in pandas is used to apply a function to every element of a DataFrame. This method is similar to the map() method, but it is applied to the whole DataFrame instead of a single column.

```python
students_df.loc[:, "FirstName":"Gender"].applymap(len) # To get the length of each text in each columns in the columns FirstName	LastName	Gender
```

```python
students_df.loc[:, "FirstName":"Gender"].apply(len) # Notice the difference between this and the one above.
```

#### 3. map()
The map() method in pandas is used to replace the values in a Series with other values. It takes a dictionary or a function as an argument and applies it to each value in the Series.

```python
# let's say our registrar entered some students name wrongly and we are asked to edit the names. 
students_df['FirstName'].map({'Alice':'Ailes', 'Bob':'Boris', 'Emily':'Esther', 'Charlie':'Stanley', 'Olivia':'Oliver', 'David':'Dave'})
# Check the output below and see that the names we did not change became `NAN` which we don't want
# So, basically if we are not changing everybody's name, we are meant to use .replace().
# Use .map() when you are changing the whole data.
```

#### 4. replace()
The replace() method in pandas is used to replace a specified value in a Series or DataFrame with another value. It can be used to replace one or more values with a single value, or to replace different values with different values. The method takes a dictionary, a list, or a scalar value as an argument.

```python
students_df['FirstName'].replace({'Alice':'Ailes', 'Bob':'Boris', 'Emily':'Esther', 'Charlie':'Stanley', 'Olivia':'Oliver', 'David':'Dave'}) 
# problem solved!!!
```

```python
# Now, we can permanently change the data!!! feels good right???
students_df['FirstName'] = students_df['FirstName'].replace({'Alice':'Ailes', 'Bob':'Boris', 'Emily':'Esther', 'Charlie':'Stanley', 'Olivia':'Oliver', 'David':'Dave'})
```

```python
students_df # Data changed
```

#### map() and replace()
Both .map() and .replace() are methods in Pandas that allow you to modify the values in a Pandas Series or DataFrame. However, there are some key differences between the two:

###### - Functionality:

.map() is used to map values from one set of values to another set of values, using either a dictionary or a function. It returns a new Series or DataFrame with the new mapped values.
.replace() is used to replace a set of values with another set of values. It can take either a dictionary or two lists as arguments. It modifies the existing Series or DataFrame in place.

###### - Usage:

.map() is typically used to apply a simple function or dictionary mapping to each element of a Series or DataFrame. For example, you could use .map() to convert a column of strings to uppercase or to replace specific values with new values using a dictionary mapping.
.replace() is typically used to replace specific values or patterns in a Series or DataFrame with new values. For example, you could use .replace() to replace all occurrences of "Male" with "M" in a column of gender values.

###### - In-place modification:

.map() does not modify the original Series or DataFrame; instead, it returns a new Series or DataFrame with the mapped values.
.replace() modifies the original Series or DataFrame in place, unless you set the inplace=True argument.
Overall, .map() and .replace() can be used to achieve similar results, but they have different use cases and should be chosen based on the specific task at hand.
