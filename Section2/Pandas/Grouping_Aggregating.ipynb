# `Grouping in numpy`
#### Grouping in pandas involves three stages
- Spliting the data
- Applying a function
- Combining the results

```python
Country_grps = Dev_survey_df.groupby('Country')
```

```python
Country_grps.get_group('Nigeria')
```

```python
Country_grps['Sexuality'].value_counts().loc['United States of America']
```

```python
Country_grps['ConvertedCompYearly'].median()
```

```python
Country_grps['ConvertedCompYearly'].agg(['median', 'mean', 'min', 'max'])
```

```python
Country_grps['ConvertedCompYearly'].max()['Canada']
```

```python
Country_grps['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Go').sum())
```

```python
Country_grps['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python'))
```

#### Challenge:
- Using pandas, Calculate the percentage of people that can code in `python` in each country

```python
NumRegistered_df = Dev_survey_df.groupby('Country')
NumRegistered_df['LanguageHaveWorkedWith'].count()
```

```python
NumCodeInPython = NumRegistered_df['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
NumCodeInPython
```

```python
Python_df = pd.concat([NumRegistered_df, NumCodeInPython])
```
