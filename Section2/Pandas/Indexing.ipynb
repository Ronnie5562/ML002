Now, loc and iloc will work with their student ids

```python
students_df.loc['1001':'1005', ['Name', 'GPA']]
```

So, if you are wondering why slicing is possible inside loc and iloc. It is because:
Both `loc` and `iloc` take in two arguments:

- Indexer for Rows: This can be a single value, list of values, slice, boolean array or callable function that returns boolean array. This parameter is used to select the rows from the DataFrame.

- Indexer for Columns: This can be a single value, list of values, slice, boolean array or callable function that returns boolean array. This parameter is used to select the columns from the DataFrame.

To reset the index back to pandas default serial numbers, use the `.reset_index(inplace=True)` method.
`inplace=True` makes the change of index permanent.
