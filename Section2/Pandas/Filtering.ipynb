```python
students_df.loc[filt, ['StudentID', 'FirstName', 'LastName', 'GPA']]
```

 The code below filters out SS3 students with a GPA higher than 3.3. 
 - `&` - AND 
 - `|` - OR
 - `~` - Negates a value. (put it in front of the filter object, to get the opposite what the filter is meant to filter out).

```python
second_filt = ((students_df['GPA'] > 3) & (students_df['Class'] == 'SS3'))
```

```python
students_df.loc[second_filt, ['StudentID', 'FirstName', 'LastName', 'GPA', 'Class']]
```

```python
new_filt = (students_df['GPA'] > 3.2)
```

```python
students_df.loc[~new_filt, ['StudentID', 'FirstName', 'LastName', 'GPA']] # Notice the tilde (~) sign in front of new_flit(the filter obj)  is used to invert boolean arrays or conditions.
```
