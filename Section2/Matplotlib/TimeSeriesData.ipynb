```python
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
plt.style.use('seaborn-v0_8')
from matplotlib import dates as mpldate
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 5, 3, 7]

plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate() # to make the dates label slant. gcf() -- get current figure
date_format = mpldates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format) # To format the date. gca() -- get current axis
plt.show()
```

#### Let's read data from a file and then plot it.

```python
data = pd.read_csv('bitcoinData.csv')
data.sort_values(by='Date', inplace=True)
price_date = data['Date']
price_close = data['Close']
# I sorted the data because the information for 17th was not in its right position.
plt.plot_date(price_date, price_close, linestyle='solid', linewidth=1)
plt.tight_layout()
plt.gcf().autofmt_xdate()
plt.show()
```
