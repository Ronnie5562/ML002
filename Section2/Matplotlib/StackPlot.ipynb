```python
from matplotlib import pyplot as plt
```

```python
Rounds = [1,2,3,4,5]
p1_scores = [10, 40, 80, 60, 90]
p2_scores = [40, 50, 60, 70, 80]
p3_scores = [20, 40, 70, 80, 100]
labels = ['player1', 'player2', 'player3']
colors = ['#0045FF', '#13FF26', '#FF4562']
plt.stackplot(Rounds, p1_scores, p2_scores, p3_scores, colors=colors, labels=labels)
plt.legend(title='LEGEND', loc='upper left')
plt.title('My Awesome Stack Plot')
plt.tight_layout()
plt.show()
```
