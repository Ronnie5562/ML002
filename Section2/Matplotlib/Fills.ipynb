```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data2.csv')
Ages = data['Age']
All_devs_salaries = data['All_Devs']
Py_devs_salaries = data['Python']
JS_devs_salaries = data['JavaScript']

plt.plot(Ages, All_devs_salaries, 'k--', label='All Devs')
plt.plot(Ages, Py_devs_salaries, label='Python Devs')
plt.fill_between(Ages, Py_devs_salaries, y2=All_devs_salaries, alpha=0.5, where=(Py_devs_salaries > All_devs_salaries), label='Above Average')
plt.fill_between(Ages,Py_devs_salaries, y2=All_devs_salaries, alpha=0.5, where=(Py_devs_salaries <= All_devs_salaries), label='Below Average')
plt.tight_layout()
plt.legend(title='Legend')
plt.xlabel('Ages')
plt.ylabel('Salaries')
plt.show()
```

```python
data = pd.read_csv('data2.csv')
Ages = data['Age']
All_devs_salaries = data['All_Devs']
Py_devs_salaries = data['Python']
JS_devs_salaries = data['JavaScript']

plt.plot(Ages, All_devs_salaries, 'k--', label='All Devs')
plt.plot(Ages, Py_devs_salaries, label='Python Devs')
plt.fill_between(Ages, Py_devs_salaries, y2=All_devs_salaries, alpha=0.5, interpolate=True,
                 where=(Py_devs_salaries > All_devs_salaries), label='Above Average')
plt.fill_between(Ages,Py_devs_salaries, y2=All_devs_salaries, alpha=0.5, interpolate=True, 
                 where=(Py_devs_salaries <= All_devs_salaries), label='Below Average')
plt.tight_layout()
plt.legend(title='Legend')
plt.xlabel('Ages')
plt.ylabel('Salaries')
plt.show()
# `where=()` is used to specify a condition that would determine where the fill will be executed.
# `alpha` - determines the opacity of the fill.
# `y2` - determines where the fill stops. The parameter before `y2` is `y1`.
# >>>>>> we dont explicitly need to write `y2=`. we can just pass its value immediately after y1.
```

Did you notice any difference between the graphs?
It is because I added `interpolation=True` to the second graph.
interpolation is a technique used to estimate the values of points on a curve between two known data points.

When we create a line plot, it connects a series of data points with straight lines, but sometimes the data points may not be evenly spaced, or there may be missing data points. In such cases, interpolation is used to estimate the values of the missing points and fill the gaps in the line.
