```python
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("seaborn-v0_8-darkgrid")
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter(x, y, s=sizes, edgecolor='black', alpha=0.9, c=colors, cmap='Blues')
cbar = plt.colorbar()
cbar.set_label('Satisfaction level')
plt.tight_layout()
plt.show()
```

##### Now, lets create a scatter plot using data from a csv file.
##### It's a data pulled from youtube on current trending videos

```python
data = pd.read_csv("data4.csv")
views = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.style.use("fivethirtyeight")
plt.scatter(views, likes, c=ratio, cmap='Reds', edgecolor="white")
plt.colorbar().set_label('Like : dislike ratio')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.tight_layout()
plt.show()
```
