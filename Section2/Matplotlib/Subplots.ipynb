```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')

data = pd.read_csv('data2.csv')
Ages = data['Age']
All_devs = data['All_Devs']
Py_devs = data['Python']
Js_devs = data['JavaScript']

plt.plot(Ages, Py_devs, label='Python')
plt.plot(Ages, Js_devs, label='JavaScript')
plt.plot(Ages, All_devs, label='All_devs')
plt.xlabel('Ages')
plt.ylabel('Programming Languages')
plt.legend()
plt.tight_layout()
plt.show()
```
