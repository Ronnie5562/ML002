```python
import csv
from collections import Counter # Counter takes in a list and it returns a dictionary with each element as a `key` and 
# the number of times the element appears in the list as its `value`
```

```python
Tlist = []
with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for line in csv_reader:
        listX = line['LanguagesWorkedWith'].split(';')
        Tlist.extend(listX)
    Counter_dict = Counter(Tlist)
languages = []
popularity = []
for item in Counter_dict.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

    
plt.bar(languages, popularity)
plt.title("Most Popular Programming Languages")
plt.xlabel("Programming Languages")
plt.ylabel("Number of developers")
# Notice that the names of the languages overlap each other in the graph below. To solve this problem, 
# we have to use an horizontal bar chart.
```

```python
# All we need to do is to add an 'h' to bar. ==> 'barh()', but we are to leave the other in which our parameters are arrange 
# Also we need to swap the x and y labels.
languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)
plt.title("Most Popular Programming Languages")
plt.xlabel("Number of developers")
plt.ylabel("Programming Languages")
```
